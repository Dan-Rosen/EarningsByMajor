# Recent College Grads from a 2010-2012 Survey
## Data is imported and rows with null values are removed

```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0])
recent_grads.describe()
```

```python
recent_grads.dropna()
```

## Scatter Plots

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median Income vs Sample Size of Major')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment Rate vs Sample Size of Major')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full Time Jobs vs Median Income of Major')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Share of Women vs Unemployment Rate of Major')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Number of Men vs Median Income of Major')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Number of Women vs Median Income of Major')
```

### Findings:
1. Students in more popular majors tend to make less money

## Histograms

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]
fig = plt.figure(figsize=(15,20))
for r in range(1, 5):
    ax = fig.add_subplot(4,2,r)
    ax = recent_grads[cols[r-1]].plot(kind='hist', rot=30)
    ax.set_title(cols[r-1])
for r in range(5, 9):
    ax = fig.add_subplot(4,2,r)
    ax = recent_grads[cols[r-1]].plot(kind='hist', rot=30)
    ax.set_title(cols[r-1])   
```

### Findings
1. There are more female dominated majors than male dominated majors
2. Most recent graduates are employed full time
3. The most common median salary is between $30,000 - $40,000
